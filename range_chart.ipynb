# Step by Step guide on how to plot Range Chart

Range Chart is a control chart for process variation, it is a variable control chart, which means that the type of observation required for it is a **continuous one**

In this section, we'll take you through on how to plot a Range chart.

```python
#Import required package.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from bs4 import BeautifulSoup
```

### Importing data to be used

```python
df = pd.read_csv('data.csv')
```

```python
#Printing the first five rows of the dataset
df.head()
```

### Next, we need to find the Range of each Sample number, i.e Difference between the Maximum observations and the minimum observations in each Sample.

```python
df['range'] = df.iloc[:,1::].max(axis = 1) - df.iloc[:,1::].min(axis = 1)
```

```python
df.head()
```

```python
#Finding the mean of the range.
r_bar = df['range'].mean()
```

### The Control Limits of R chart calculated as follows:
- UCL(upper control limit) = r_bar * D4
- CL (control limit) = r_bar
- LCL(lower control limit) =  r_bar * D3

where:

    r_bar = The mean of Range

The D4 and D3 values depends on the **number of observations per Sample number(m)** , in this case my **m = 5**, so we're to check for D3 and D4 when **m = 5** , the quality control table to check for R constant can be found <a href='https://www.mymathtables.com/statistic/xbar-rchart-table.html'>here</a>

Alternatively, let's we can also webscrape it here:


```python
#using requests to connect to the webpage
req = requests.get('https://www.mymathtables.com/statistic/xbar-rchart-table.html')
```

```python
soup = BeautifulSoup(req.text,'lxml')
```

```python
#Webscraping the table
table = pd.read_html(str(soup.find('table')),flavor = 'lxml')[0]
```

```python
table
```

### From the table above, we can see that the corresponding value for D3 and D4 when is m is 5 is 0 and 2.114, we can now find UCL and LCL for Range chart

```python
D3 = 0
print('D3 value is: ',D3)
D4 = 2.114
print('D4 value is: ',D4)

UCL = D4 * r_bar
print('UCL value is: ',UCL)
LCL = D3 * r_bar
print('LCL value is: ',LCL)
CL = r_bar 
print('CL value is: ',CL)
```

### Now, let's plot the Chart:

```python
plt.figure(figsize = [10,6])
sns.lineplot(data = df, x = 'sample_number', y = 'range', marker = 'o')
sns.lineplot(data = df, x = 'sample_number', y = UCL,linestyle = 'dashed')
sns.lineplot(data = df, x = 'sample_number', y =  LCL,linestyle = 'dashed')
sns.lineplot(data = df, x = 'sample_number', y = CL,linestyle = 'dashed')
plt.text(df.sample_number.max() + 1.5,LCL, 'LCL = ' + str(round(LCL,2)))
plt.text(df.sample_number.max() + 1.5,UCL,'UCL = ' + str(round(UCL,2)))
plt.text(df.sample_number.max() + 1.5,CL,'CL =' + str(round(CL,2)))
plt.title('Range Chart',fontsize = 16);
plt.xticks(range(min(df['sample_number']),max(df['sample_number'])+1));
plt.show()
```

Since all data points falls within the UCL and LCL, we can say that the process variability is in control
